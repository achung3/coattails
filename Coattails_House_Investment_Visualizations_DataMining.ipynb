```python
#import read/fetch csv file with all the transcations for the House Reps involved in documented financial gains from https://housestockwatcher.com/api as pandas dataframe
df = pd.read_csv("all_transactions_house.csv")
#call the data frame not printed so it retains the correct organization of rows columns
df
```

```python
# filters data to only show the most recently recorded completed year - 2021
df = df[df['disclosure_year'] == 2021]
df
```

```python
#remove uncessecary data for the particular visualization
df = df.drop(columns=['transaction_date', 'disclosure_date', 'owner', 'type', 'district', 'ptr_link'])
df
```

```python
#Only show the financial investments of the House reps with significant gain - " good buys"
df = df[df['cap_gains_over_200_usd'] == True]
df
```

```python
#counts frequency of asset within the list
dfhg.value_counts(subset=None, sort=True, ascending=False)
```

```python
#turns the counter into a dictionary with named ticker value and the frequency number in the pair
#This dict will be used for the FREQUENCY GRAPH
from collections import Counter
my_dict = dict(Counter(dfhg.ticker))
my_dict
```

```python
#turns the dictionary into a dataframe 
dfhgc = pd.DataFrame(list(my_dict.items()), columns = ['ticker', 'frequency'])
dfhgc
```

```python
#Sort Data in from greatest frequency to least, meaning ticker representing the profitable financial investment asset that was the most bought is at the top
dfhgcc = dfhgc.sort_values(by='frequency', ascending=False)
dfhgcc
```

```python
#export dfhgcc as house_good_Investments_Sorted.csv
#TO BE imported and USED FOR px.bar figure
dfhgcc.to_csv(r"C:\YourPATH\house_Good_Investments_Sorted.csv", index=False)
```
